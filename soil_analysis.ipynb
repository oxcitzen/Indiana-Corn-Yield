```python
columns = ['Drainage', 'texture_primary',
       'texture_secondary', 'stratified', 'rvindicator_tex',
       'RoadwayRelevanceSoilGrade', 'HumanErosion','Drainage_class', 'Flood_freq']
def plot_frequency_grid(df, columns, rows=3, cols=3):
    """
    Creates a grid of frequency plots for specified columns in the DataFrame.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the data.
        columns (list): List of column names to create frequency plots for.
        rows (int): Number of rows in the grid.
        cols (int): Number of columns in the grid.

    Returns:
        A matplotlib figure containing the grid of frequency plots.
    """
    # Filter out the specified columns
    plot_data = df[columns]
    
    # Create the subplots
    fig, axes = plt.subplots(rows, cols, figsize=(20, 15))
    axes = axes.flatten()  # Flatten to 1D array for easy iteration

    for i, col in enumerate(plot_data.columns):
        if i >= rows * cols:  # Check if we have more columns than grid slots
            break
        sns.histplot(plot_data[col], kde=True, ax=axes[i])
        axes[i].set_title(f'Frequency of {col}', fontsize=15)
        axes[i].set_xlabel('')
        axes[i].set_ylabel('')
    
    plt.tight_layout()
    plt.show()

plot_frequency_grid(data, columns)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor

# Combine the training and validation data
X_train_val = np.concatenate((X_train, X_val), axis=0)
y_train_val = np.concatenate((y_train, y_val), axis=0)

# Define your Random Forest Regressor model
rf_model = RandomForestRegressor(**best_params, random_state=42)

# Specify the number of folds for cross-validation
num_folds = 5
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# List to store cross-validation scores
cv_scores = []

# Perform cross-validation
for train_index, test_index in kf.split(X_train_val):
    X_train_fold, X_test_fold = X_train_val[train_index], X_train_val[test_index]
    y_train_fold, y_test_fold = y_train_val[train_index], y_train_val[test_index]
    
    # Fit the model on the training fold
    rf_model.fit(X_train_fold, y_train_fold)
    
    # Predict on the test fold
    y_pred_fold = rf_model.predict(X_test_fold)
    
    # Calculate mean squared error for this fold
    fold_mse = mean_squared_error(y_test_fold, y_pred_fold)
    
    # Append the fold's MSE to the list of cross-validation scores
    cv_scores.append(fold_mse)

# Print the cross-validation scores
print("Cross-Validation Scores (MSE):", cv_scores)

# Calculate the mean and standard deviation of the cross-validation scores
mean_cv_score = np.mean(cv_scores)
std_cv_score = np.std(cv_scores)

print("Mean Cross-Validation Score (MSE):", mean_cv_score)
print("Standard Deviation of Cross-Validation Scores (MSE):", std_cv_score)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Initialize the gradient boosted regressor
gb_model = GradientBoostingRegressor()

# Fit the model on the training data
gb_model.fit(X_train, y_train)

# Evaluate the model on the validation set
validation_score = gb_model.score(X_val, y_val)
print("Validation Score:", validation_score)

# Evaluate the model on the test set
test_score = gb_model.score(X_test, y_test)
print("Test Score:", test_score)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Make predictions on the validation set
y_val_pred = gb_model.predict(X_val)

# Calculate additional evaluation metrics
mae = mean_absolute_error(y_val, y_val_pred)
mse = mean_squared_error(y_val, y_val_pred)
rmse = mean_squared_error(y_val, y_val_pred, squared=False)  # squared=False for RMSE
r2 = r2_score(y_val, y_val_pred)

# Print the additional evaluation metrics
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared Score:", r2)
```

```python
import matplotlib.pyplot as plt

# Calculate residuals
residuals = y_val - y_val_pred

# Create residual plot
plt.figure(figsize=(8, 6))
plt.scatter(y_val_pred, residuals, color='blue', alpha=0.5)
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='--')  # Add horizontal line at y=0
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
import numpy as np

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}

# Initialize the gradient boosted regressor
gb_model = GradientBoostingRegressor()

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=gb_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)

# Perform grid search
grid_search.fit(X_train_val, y_train_val)

# Print the best parameters found
print("Best Parameters:", grid_search.best_params_)

# Get the best model
best_gb_model = grid_search.best_estimator_
#validation score
# Evaluate the model on the validation set
validation_score = best_gb_model.score(X_val, y_val)
print("Validation Score:", validation_score)

# Evaluate the best model on the test set
test_score = best_gb_model.score(X_test, y_test)
print("Test Score:", test_score)
```
